# 🎬 KinoVibe — Сборка APK в Google Colab

**Инструкция:**
1. Runtime → Run all
2. Дождись завершения (~10-15 минут)
3. Скачай APK в последней ячейке

```
# Шаг 1: Системные зависимости
!apt-get update -q
!apt-get install -y -q curl git unzip xz-utils zip libglu1-mesa openjdk-17-jdk
print('✅ Системные зависимости установлены')
```

```
# Шаг 2: Flutter SDK
import os
!git clone https://github.com/flutter/flutter.git -b stable /opt/flutter --depth 1 -q
os.environ['PATH'] = f"/opt/flutter/bin:{os.environ['PATH']}"
!flutter --version
print('✅ Flutter установлен')
```

```
# Шаг 3: Android SDK (cmdline-tools)
!mkdir -p /opt/android-sdk/cmdline-tools
!wget -q https://dl.google.com/android/repository/commandlinetools-linux-11076708_latest.zip -O /tmp/cmdtools.zip
!unzip -q /tmp/cmdtools.zip -d /opt/android-sdk/cmdline-tools
!mv /opt/android-sdk/cmdline-tools/cmdline-tools /opt/android-sdk/cmdline-tools/latest
os.environ['ANDROID_SDK_ROOT'] = '/opt/android-sdk'
os.environ['PATH'] = f"/opt/android-sdk/cmdline-tools/latest/bin:{os.environ['PATH']}"
!yes | sdkmanager --licenses > /dev/null 2>&1
!sdkmanager 'platform-tools' 'platforms;android-34' 'build-tools;34.0.0' > /dev/null 2>&1
print('✅ Android SDK установлен')
```

```
# Шаг 4: Клонируем репозиторий KinoVibe
REPO_URL = 'https://github.com/vaalchik-lang/KINOVIBE.git'  # замени если нужно
!git clone {REPO_URL} /content/KINOVIBE -q
%cd /content/KINOVIBE/client
print('✅ Репозиторий клонирован')
```

```
# Шаг 5: Сборка APK
!flutter pub get
!flutter build apk --release
print('✅ APK собран')
```

```
# Шаг 6: Скачать APK
from google.colab import files
import os
apk_path = 'build/app/outputs/flutter-apk/app-release.apk'
if os.path.exists(apk_path):
    size = os.path.getsize(apk_path) / 1024 / 1024
    print(f'📦 APK готов: {size:.1f} MB')
    files.download(apk_path)
else:
    print('❌ APK не найден — проверь логи выше')
```
